# Model Training

````python
x = df_cancer.drop(['target'], axis= 1)
````

````python
x
````

````python
y = df_cancer['target']
````

````python
y
````

````python
# Splitting the data into training and testing
````

````python
from sklearn.model_selection import train_test_split
````

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
````

````python
x_train
````

````python
y_train
````

````python
x_test
````

````python
y_test
````

````python
#Everything looks great
````

````python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
````

````python
svc_model = SVC()
````

````python
svc_model.fit(x_train, y_train)
````

````python
# Model Evaluation
````

````python
y_predict = svc_model.predict(x_test)
````

````python
y_predict
````

````python
cm = confusion_matrix(y_test, y_predict)
````

````python
cm
````

````python
sns.heatmap(cm, annot = True)
````

# Improving The model


````python
# Data Nomarlization
````

````python
# we want all our data to be in between 0 and 1
# we can do this by feature scaling
````

````python
min_train = x_train.min()
range_train = (x_train - min_train).max()
x_train_scaled = (x_train - min_train) / range_train
````

````python
sns.scatterplot(x = x_train['mean area'], y = x_train['mean smoothness'], hue = y_train)
````

````python
sns.scatterplot(x = x_train_scaled['mean area'], y = x_train_scaled['mean smoothness'], hue = y_train)
````

````python
# data is similar but the range is different, ie normalized
````

````python
min_test = x_test.min()
range_test = (x_test - min_test).max()
x_test_scaled = (x_test - min_test) / range_test
````

````python
svc_model.fit(x_train_scaled, y_train)
````

````python
y_predict = svc_model.predict(x_test_scaled)
````

````python
cm = confusion_matrix(y_test, y_predict)
````

````python
cm
````

````python
sns.heatmap(cm, annot = True)
````

````python
print(classification_report(y_test, y_predict))
````

````python
# Improving the C and Gamma parameters of the SVM
#sklearn searches for the best parameter for us
````

````python
param_grid = {'C': [0.1, 1, 10, 100], 'gamma':[1, 0.1, 0.01, 0.001], 'kernel':['rbf']}
````

````python
from sklearn.model_selection import GridSearchCV
````

````python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose =5)
````

````python
grid.fit(x_train_scaled, y_train)
````

````python
grid.best_params_
````

````python
grid_prediction = grid.predict(x_test_scaled)
````

````python
cm = confusion_matrix(y_test, grid_prediction)
````

````python
cm
````

````python
sns.heatmap(cm, annot = True)
````

````python
print(classification_report(y_test, grid_prediction))
````
